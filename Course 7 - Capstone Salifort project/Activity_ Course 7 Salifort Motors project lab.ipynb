Now, examine variables that you're interested in, and create plots to visualize relationships between variables in the data.

```python
# Set plot style
sns.set(style='whitegrid')

# Define custom palette for 'left_company'
palette = {0: '#4c72b0', 1: '#dd8452'}  # blue for stayed, orange for left

# Set up the figure with 2 subplots: boxplot and histogram
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# --- Boxplot: average_montly_hours by num_projects, grouped by left_company ---
sns.boxplot(ax=axes[0],
            data=df1,
            x='num_projects',
            y='average_montly_hours',
            hue='left_company',
            palette=palette)
axes[0].set_title('Average Monthly Hours by Number of Projects')
axes[0].set_xlabel('Number of Projects')
axes[0].set_ylabel('Average Monthly Hours')

# Fix legend to use proper labels and colors
handles, labels = axes[0].get_legend_handles_labels()
axes[0].legend(handles, ['Stayed', 'Left'], title='Left Company')

# --- Histogram: num_projects distribution by left_company ---
sns.histplot(ax=axes[1],
             data=df1,
             x='num_projects',
             hue='left_company',
             multiple='dodge',
             shrink=0.8,
             discrete=True,
             palette=palette)
axes[1].set_title('Distribution of Number of Projects')
axes[1].set_xlabel('Number of Projects')
axes[1].set_ylabel('Count')

# Fix legend to use proper labels and colors
handles, labels = axes[1].get_legend_handles_labels()
axes[1].legend(handles, ['Stayed', 'Left'], title='Left Company')

# Adjust layout and display
plt.tight_layout()
plt.show()
```

It might be natural that people who work on more projects also tend to work longer hours. This appears to be the case here, with the mean hours for each group (stayed and left) increasing as the number of projects worked rises. However, a few observations stand out from this plot.

1. There are two groups of employees who left the company:
   **(A)** Those who worked considerably less than their peers with the same number of projects, and
   **(B)** Those who worked much more.

   Among those in group A, it’s possible they were let go, or they may have already given notice and were assigned fewer hours on their way out. For group B, it's reasonable to infer they likely quit voluntarily. These employees probably carried a heavier workload and may have been key contributors to their projects.

2. Notably, **everyone with seven projects left the company**, and the interquartile ranges of this group — as well as those who left with six projects — were roughly **255–295 hours per month**, substantially more than any other group.

3. The optimal number of projects for employees to work on seems to be **4–5**, as the ratio of employees who left to those who stayed is very low in this cohort.

4. Assuming a **40-hour workweek** and **2 weeks of vacation** per year, the average number of working hours per month for an employee working Monday through Friday is:

$$
\frac{(52 - 2) \times 40}{12} = \frac{50 \times 40}{12} = 166.67 \text{ hours per month}
$$

This means that, aside from employees working on two projects, every group—even those who stayed—worked considerably more hours than this. It suggests that employees at this company are generally **overworked**.

### Identify the types of models most appropriate for this task.

For predicting whether an employee will leave the company—a binary classification problem—several model types are appropriate:

1. **Logistic Regression:**

   * Well-suited for binary outcomes.
   * Easy to interpret, allowing HR to understand feature impacts on employee turnover.
   * Assumes a linear relationship between predictors and the log-odds of the outcome.

2. **Decision Trees:**

   * Handle non-linear relationships and interactions between variables well.
   * Robust to outliers and do not require data scaling.

3. **Random Forests / Gradient Boosting Machines:**

   * Ensemble methods that often achieve higher accuracy by combining multiple decision trees.
   * Can handle complex patterns and feature interactions, but less interpretable than logistic regression.

4. **Support Vector Machines (SVM):**

   * Effective in high-dimensional spaces, can handle non-linear boundaries with kernels.
   * Typically require tuning and may be less interpretable.

5. **Neural Networks:**

   * Powerful for capturing complex patterns but may be overkill for this dataset and require more tuning and data.

Considering the project’s emphasis on interpretability and actionable insights for HR, **Logistic Regression** is a strong starting point. If improved predictive performance is needed, ensemble tree-based methods could be explored in subsequent iterations.

```python
# Define your hyperparameter grid
cv_params = {
    'max_depth': [4, 6, 8, None],
    'min_samples_leaf': [2, 5, 1],
    'min_samples_split': [2, 4, 6]
}

# Define scoring dictionary with all metrics
scoring = {
    'AUC': 'roc_auc',
    'precision': 'precision',
    'recall': 'recall',
    'f1': 'f1',
    'accuracy': 'accuracy'
}

# Instantiate Decision Tree classifier
tree = DecisionTreeClassifier(random_state=0)

# Instantiate GridSearchCV with multiple scoring metrics
grid_search = GridSearchCV(
    estimator=tree,
    param_grid=cv_params,
    scoring=scoring,
    refit='AUC',   # Use AUC to select best model
    cv=5
)

# Fit model on training data (replace X_train, y_train with your data)
grid_search.fit(X_train, y_train)

# Function to extract all scores from the GridSearchCV results
def make_results(model_name: str, model_object, metric: str):
    '''
    Arguments:
        model_name (str): model name for display
        model_object: fitted GridSearchCV object
        metric (str): main metric to pick the best model ('auc', 'precision', 'recall', 'f1', 'accuracy')

    Returns:
        DataFrame with precision, recall, f1, accuracy, and auc for best model
    '''
    metric_dict = {
        'auc': 'mean_test_AUC',
        'precision': 'mean_test_precision',
        'recall': 'mean_test_recall',
        'f1': 'mean_test_f1',
        'accuracy': 'mean_test_accuracy'
    }

    # Convert cv_results_ to DataFrame
    cv_results = pd.DataFrame(model_object.cv_results_)

    # Find index of best model according to chosen metric
    best_index = cv_results[metric_dict[metric]].idxmax()

    # Extract results for best model
    best_results = cv_results.loc[best_index]

    # Create a DataFrame with all relevant scores
    results_table = pd.DataFrame({
        'model': [model_name],
        'precision': [best_results['mean_test_precision']],
        'recall': [best_results['mean_test_recall']],
        'F1': [best_results['mean_test_f1']],
        'accuracy': [best_results['mean_test_accuracy']],
        'auc': [best_results['mean_test_AUC']]
    })

    return results_table

# Use the function to get all CV scores, using 'auc' as the main metric
tree_cv_results = make_results('Decision Tree CV', grid_search, 'auc')

tree_cv_results
```

Define a function that gets all the scores from a model's predictions.

```python
def get_scores(model_name:str, model, X_test_data, y_test_data):
    '''
    Generate a table of test scores.

    In: 
        model_name (string):  How you want your model to be named in the output table
        model:                A fit GridSearchCV object
        X_test_data:          numpy array of X_test data
        y_test_data:          numpy array of y_test data

    Out: pandas df of precision, recall, f1, accuracy, and AUC scores for your model
    '''

    preds = model.best_estimator_.predict(X_test_data)

    auc = roc_auc_score(y_test_data, preds)
    accuracy = accuracy_score(y_test_data, preds)
    precision = precision_score(y_test_data, preds)
    recall = recall_score(y_test_data, preds)
    f1 = f1_score(y_test_data, preds)

    table = pd.DataFrame({'model': [model_name],
                          'precision': [precision], 
                          'recall': [recall],
                          'f1': [f1],
                          'accuracy': [accuracy],
                          'AUC': [auc]
                         })
  
    return table
```

## Recall evaluation metrics

- **AUC** is the area under the ROC curve; it's also considered the probability that the model ranks a random positive example more highly than a random negative example.
- **Precision** measures the proportion of data points predicted as True that are actually True, in other words, the proportion of positive predictions that are true positives.
- **Recall** measures the proportion of data points that are predicted as True, out of all the data points that are actually True. In other words, it measures the proportion of positives that are correctly classified.
- **Accuracy** measures the proportion of data points that are correctly classified.
- **F1-score** is an aggregation of precision and recall.
